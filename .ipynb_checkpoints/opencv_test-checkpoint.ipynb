```python
import cv2
from cv2 import matchTemplate
import numpy as np
import hashlib
import base64
import matplotlib.pyplot as plt
```

```python
#Testing SHA1 hashing
tag = 'https://upload.wikimedia.org/wikipedia/commons/thumb/1/1d/Harvey_2017_track.png/300px-Harvey_2017_track.png'
result = hashlib.sha1(tag.encode()).hexdigest()
print(result)
```

```python
#Testing matchTemplate
image = cv2.imread('images/target-images/hurricane-harvey/hh_shark.jpg')
img = image.copy()
print(len(img), len(img[0]), img.shape)

target = cv2.imread('images/target-images/hurricane-harvey/hh_shark_cropped.jpg')
print(len(target), len(target[0]), target.shape)

w, h = target.shape[0:2]
print('Target Shape:', w, h)

res = cv2.matchTemplate(img, target, cv2.TM_CCOEFF)
print(res.shape, len(res), len(res[0]))
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)
print(cv2.minMaxLoc(res))


top_left = max_loc
print('Top Left Corner:', top_left)

bottom_right = (top_left[0] + w, top_left[1] + h)
print('Bottom Right Corner:', bottom_right)

print(img)
cv2.rectangle(img, top_left, bottom_right, 255, thickness=2)

plt.subplot(121)
plt.imshow(res)
plt.title('Matching Result')
plt.xticks([])
plt.yticks([])
plt.subplot(122)
plt.imshow(img)
plt.title('Detected Point')
plt.xticks([])
plt.yticks([])

plt.show()

```
